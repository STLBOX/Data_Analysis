```python
import pandas as pd
import numpy as np
import os
```

```python
def get_report(i):
    doc_num = i
    path = os.path.join("D:\wind数据下载", doc_list[doc_num])
    report = pd.read_excel(path)
    return report

def get_number_institution(r0, get_rep):
    # 判断年报的长度
    start = r0.index(get_rep)
    number = 1
    offset = 2
    while (r0[start + offset] == number) & ((start + offset) < len(r0) -1):
        offset += 1
        number += 1
    return start + 2, number - 1

# 获得数据,3k多个excel文档
doc_list = os.listdir(r"D:\wind数据下载")

for get_rep in  ["2011年报", "2012年报", "2013年报", "2014年报", "2015年报", "2016年报"]:
    timer = 1
    for i in range(len(doc_list)):
        report = get_report(i)
        doc_str = doc_list[i]
        try:
            num = doc_str.index("[")
            name = doc_str[:num]
            code = doc_str[num+1: num+7]
        except:
            continue

        r0 = report.iloc[:,0].to_list()
        # 存在某个时期的年报，就加入到新的excel中
        if get_rep in r0:
            print(name,i)
            start, number_institutions = get_number_institution(r0, get_rep)
            sub_report = report.iloc[start:start+number_institutions,1:5]
            # 修改columns
            sub_report.columns = ["机构名称", "机构类型", "合并数量(万股)", "占流通A股比例(%)"]
            sub_report.insert(0, "上市公司名称", name)
            sub_report.insert(1, "代码", code)

            if timer == 1:
                new_report = sub_report
                timer = -1
            else:
                new_report = new_report.append(sub_report, ignore_index=True)
    print(len(new_report),get_rep)
    new_report.to_excel("{}.xlsx".format(get_rep), index=False)
```
